# Dasar Text-Preprocessing dengan Python#
Text data needs to be cleaned and encoded to numerical values before giving them to machine learning models, this process of cleaning and encoding is called as text preprocessing.

## Case Folding: Lowercase ##
Merubah format teks menjadi format huruf kecil semua (_lowercase_).

```python
kalimat = "Pemerintah Indonesia dan Amerika Serikat sepakat menurunkan tarif barang asal Indonesia ke AS menjadi 19%, sementara Indonesia menghapus 99% hambatan tarif bagi produk-produk AS."
lower_case = kalimat.lower()
print(lower_case)
```

## Case Folding: Removing Number ##
Menghapus karakter angka.

```python
import re

kalimat = "Pemerintah Indonesia dan Amerika Serikat sepakat menurunkan tarif barang asal Indonesia ke AS menjadi 19%, sementara Indonesia menghapus 99% hambatan tarif bagi produk-produk AS."
hasil = re.sub(r"\d+", "", kalimat)
print(hasil)
```

## Case Folding: Removing Punctuation ##
Menghapus karakter tanda baca.

```python
import string

kalimat = "Kesepakatan ini diteken oleh Menko Perekonomian Airlangga Hartarto dan Perwakilan Perdagangan AS Jamieson Greer, pada Jumat (20/02) pagi WIB, menyusul negosiasi selama berbulan-bulan."
hasil = kalimat.translate(__builtins__.str.maketrans("","",string.punctuation))
print(hasil)
```

## Case Folding: Removing Whitespace ##
Menghapus karakter kosong.

```python
kalimat = " Kesepakatan ini diteken oleh Menko Perekonomian Airlangga Hartarto dan Perwakilan Perdagangan AS Jamieson Greer, pada Jumat (20/02) pagi WIB, menyusul negosiasi selama berbulan-bulan. "
hasil = kalimat.strip()
print(hasil)
```

## Tokenizing: Word Tokenizing Using NLTK Module ##
Menggunakan _library_ NLTK untuk memisahkan kata dalam sebuah kalimat.


```python
from nltk.tokenize import word_tokenize

kalimat = "Rincian perjanjian tidak disertakan dalam pernyataan Hartarto dan pejabat terkait."

tokens = nltk.tokenize.word_tokenize(kalimat)
print(tokens)
```

## Tokenizing with Case Folding ##
Menggabungkan teknik _Case Foling_ dengan _Tokenizing_.

```python
from nltk.tokenize import word_tokenize

kalimat = "Rincian perjanjian tidak disertakan dalam pernyataan Hartarto dan pejabat terkait."
kalimat = kalimat.translate(__builtins__.str.maketrans('','',string.punctuation)).lower()

tokens = nltk.tokenize.word_tokenize(kalimat)
print(tokens)
```

## Frequency Distribution ##
Menghitung frekuensi kemunculan setiap tokens(kata) dalam teks.

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

kalimat = "Beberapa hal yang dilakukan, antara lain: memberikan masukan konkret dari industri di lapangan, menginventarisasi hambatan yang dihadapi pelaku usaha, serta membangun komunikasi dengan kolega dunia usaha di AS agar bisa mendorong terwujudnya kesepakatan bisnis yang riil dan saling menguntungkan. "
kalimat = kalimat.translate(__builtins__.str.maketrans('','',string.punctuation)).lower()

tokens = nltk.tokenize.word_tokenize(kalimat)
kemunculan = nltk.FreqDist(tokens)
print(kemunculan.most_common())
```

## Frequency Distribution Visualization with Matplotlib ##
Untuk menggambarkan frekuensi kemunculan setiap tokens dapat menggunakan _library_ __matplotlib__ pada Python.

https://matplotlib.org

```python
import matplotlib.pyplot as plt

kemunculan.plot(30,cumulative=False)
plt.show()
```

## Tokenizing: Sentences Tokenizing Using NLTK Module ##
Memisahkan kalimat dalam sebuah paragraf.

```python
from nltk.tokenize import sent_tokenize

kalimat = "Surplus neraca dagang ini, lanjut Telisa, merupakan salah satu sumber devisa. Ia menjelaskan besaran devisa sempat melemah pada periode semester lalu karena harga komoditas yang menjadi andalan seperti sawit sedang stagnan. "

tokens = nltk.tokenize.sent_tokenize(kalimat)
print(tokens)
```

## Filtering using NLTK ##

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

kalimat = "Surplus neraca dagang ini, lanjut Telisa, merupakan salah satu sumber devisa. Ia menjelaskan besaran devisa sempat melemah pada periode semester lalu karena harga komoditas yang menjadi andalan seperti sawit sedang stagnan. "
kalimat = kalimat.translate(__builtins__.str.maketrans('','',string.punctuation)).lower()

tokens = word_tokenize(kalimat)
listStopword =  set(stopwords.words('indonesian'))

removed = []
for t in tokens:
    if t not in listStopword:
        removed.append(t)

print(removed)
```

## Filtering using Sastrawi ##

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from nltk.tokenize import word_tokenize

factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

kalimat = "Surplus neraca dagang ini, lanjut Telisa, merupakan salah satu sumber devisa. Ia menjelaskan besaran devisa sempat melemah pada periode semester lalu karena harga komoditas yang menjadi andalan seperti sawit sedang stagnan. "
kalimat = kalimat.translate(__builtins__.str.maketrans('','',string.punctuation)).lower()

stop = stopword.remove(kalimat)
tokens = nltk.tokenize.word_tokenize(stop)

print(tokens)
```

## Add Custom Stopword ##
Menambahkan kata di _stopword_ untuk dihilangkan pada sebuah teks.

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
from nltk.tokenize import word_tokenize

# ambil stopword bawaan
stop_factory = StopWordRemoverFactory().get_stop_words()
more_stopword = ['neraca', 'besaran']

kalimat = "Surplus neraca dagang ini, lanjut Telisa, merupakan salah satu sumber devisa. Ia menjelaskan besaran devisa sempat melemah pada periode semester lalu karena harga komoditas yang menjadi andalan seperti sawit sedang stagnan. "
kalimat = kalimat.translate(__builtins__.str.maketrans('','',string.punctuation)).lower()

# menggabungkan stopword
data = stop_factory + more_stopword

dictionary = ArrayDictionary(data)
str = StopWordRemover(dictionary)
tokens = nltk.tokenize.word_tokenize(str.remove(kalimat))

print(tokens)
```

## Stemming : Porter Stemming Algorithm using NLTK ##

https://tartarus.org/martin/PorterStemmer/index-old.html

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

kata = ["program", "programs", "programer", "programing", "programers"]

for k in kata:
    print(k, " : ", ps.stem(k))
```

Tahapan Utama Text Preprocessing

Proses pembersihan dan pengolahan data teks ini mencakup beberapa teknik utama:
1. Case Folding

Tahap awal untuk menyeragamkan format teks agar lebih mudah diolah.

    Lowercase: Mengubah semua huruf menjadi huruf kecil.

    Removing Numbers: Menghapus karakter angka menggunakan Regular Expression (re).

    Removing Punctuation: Menghapus tanda baca agar tidak dianggap sebagai bagian dari kata.

    Removing Whitespace: Menghapus spasi kosong yang tidak perlu di awal atau akhir kalimat.

2. Tokenizing

Proses memecah teks menjadi potongan-potongan yang lebih kecil (token).

    Word Tokenizing: Memecah kalimat menjadi kata-kata tunggal menggunakan library NLTK.

    Sentence Tokenizing: Memisahkan paragraf menjadi kalimat-kalimat penyusunnya.

    Frequency Distribution: Menghitung frekuensi kemunculan setiap kata dan memvisualisasikannya dalam bentuk grafik menggunakan Matplotlib.

3. Filtering (Stopword Removal)

Proses membuang kata-kata yang sering muncul namun memiliki sedikit makna (seperti "yang", "dan", "di").

    Dapat dilakukan menggunakan daftar kata dari library NLTK atau Sastrawi.

    Mendukung penambahan Custom Stopwords jika ada kata-kata spesifik yang ingin dihilangkan dari hasil analisis.

4. Stemming

Proses mengubah kata berimbuhan menjadi kata dasarnya.

    Porter Stemming: Digunakan untuk teks berbahasa Inggris melalui NLTK (contoh: "programming" menjadi "program").

    Sastrawi Stemmer: Digunakan khusus untuk teks Bahasa Indonesia agar kata seperti "merupakan" kembali ke kata dasar "rupa".

Library Utama yang Digunakan:

    NLTK: Untuk tokenizing, pembersihan tanda baca, dan filtering umum.

    Sastrawi: Library khusus untuk menangani stemming dan stopword dalam Bahasa Indonesia.

    Matplotlib: Untuk visualisasi data hasil pemrosesa

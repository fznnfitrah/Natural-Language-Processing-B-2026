# **Lemmatizer**

```python
!pip install nlp-id
```

```python
from nlp_id.lemmatizer import Lemmatizer
lemmatizer = Lemmatizer()
lemmatizer.lemmatize('  Pemerintah melalui Kementerian Komunikasi dan Digital (Komdigi) resmi menunda akses anak di bawah usia 16 tahun ke sejumlah platform digital berisiko tinggi.')
```

# Kesimpulan

Kode ini menunjukkan bagaimana menggunakan pustaka "nlp-id", khususnya komponen Lemmatizer, untuk melakukan lemmatisasi pada sebuah kalimat Bahasa Indonesia. Lemmatisasi adalah proses mengurangi kata-kata ke bentuk dasar atau akar katanya. Seperti yang ditunjukkan oleh output, kalimat masukan telah berhasil dilematisasi.


# **Tokenizer**

```python
!pip install nlp-id

import nltk
nltk.download('punkt_tab')

from nlp_id.tokenizer import PhraseTokenizer
```

## **The Standart Tokens**

```python
from nlp_id.tokenizer import Tokenizer
tokenizer = Tokenizer()
tokenizer.tokenize('  Pemerintah melalui Kementerian Komunikasi dan Digital (Komdigi) resmi menunda akses anak di bawah usia 16 tahun ke sejumlah platform digital berisiko tinggi.')
```

## **The Phrase Tokens**

```python
tokenizer = PhraseTokenizer()
tokenizer.tokenize('  Pemerintah melalui Kementerian Komunikasi dan Digital (Komdigi) resmi menunda akses anak di bawah usia 16 tahun ke sejumlah platform digital berisiko tinggi.')
```

## Kesimpulan

Kode ini bertujuan untuk mendemonstrasikan dua jenis tokenisasi teks bahasa Indonesia menggunakan pustaka nlp-id:

  * Tokenisasi Standar: Menggunakan Tokenizer untuk memisahkan kalimat menjadi kata-kata tunggal.

  * Tokenisasi Frasa: Menggunakan PhraseTokenizer untuk mengenali dan mempertahankan frasa multi-kata (seperti 'Kementerian Komunikasi' atau 'platform digital') sebagai satu unit token.

Secara keseluruhan, kode ini menunjukkan bagaimana nlp-id dapat digunakan untuk memecah teks menjadi unit-unit yang bermakna, baik itu kata individu maupun frasa.

# **Postag**

```python
!pip install nlp-id

import nltk
nltk.download('punkt_tab')

from nlp_id.postag import PosTag
```

## **Standard POS Tagger**

```python
postagger = PosTag()
postagger.get_pos_tag('Pemerintah melalui Kementerian Komunikasi dan Digital (Komdigi) resmi menunda akses anak di bawah usia 16 tahun ke sejumlah platform digital berisiko tinggi.')
```

## **Phrase POS Tagger**

```python
postagger.get_phrase_tag('  Pemerintah melalui Kementerian Komunikasi dan Digital (Komdigi) resmi menunda akses anak di bawah usia 16 tahun ke sejumlah platform digital berisiko tinggi.')
```

# Kesimpulan

Kode yang Anda berikan menunjukkan penggunaan pustaka nlp-id untuk tugas pemrosesan bahasa alami dalam Bahasa Indonesia. Secara khusus, ini menunjukkan dua fungsionalitas:

  * Penandaan Part-of-Speech (POS) Standar: Contoh pertama menggunakan postagger.get_pos_tag untuk memecah kalimat menjadi kata-kata individual dan menetapkan setiap kata peran tata bahasa yang sesuai (misalnya, Kata Benda, Kata Kerja, Kata Sifat). Hasilnya adalah daftar tupel, di mana setiap tupel berisi kata dan tag POS-nya.

  * Penandaan Frasa POS: Contoh kedua menggunakan postagger.get_phrase_tag yang mirip dengan penandaan POS standar tetapi juga mengidentifikasi frasa umum dalam kalimat (misalnya, Frasa Kata Benda (NP), Frasa Kata Kerja (VP), Frasa Kata Sifat (ADJP)). Ini memberikan tingkat analisis linguistik yang sedikit lebih tinggi, mengelompokkan kata-kata terkait ke dalam frasa sambil tetap memberikan tag kata individual yang sesuai.

Pada intinya, kode ini menyediakan cara cepat untuk melakukan analisis sintaksis dasar pada teks Bahasa Indonesia menggunakan pustaka nlp-id.

# **Stopword**

```python
!pip install nlp-id

from nlp_id.stopword import StopWord
stopword = StopWord()
stopword.get_stopword()
```

```python
text = "Komdigi menyatakan proses penonaktifan akun tersebut akan dilakukan secara bertahap hingga seluruh platform mematuhi kewajiban yang diatur dalam regulasi."
stopword = StopWord()
stopword.remove_stopword(text)
```

```python
paragraph = "Pemerintah melalui Kementerian Komunikasi dan Digital (Komdigi) resmi menunda akses anak di bawah usia 16 tahun ke sejumlah platform digital berisiko tinggi."
stopword.remove_stopword(paragraph)
```

# Kesimpulan

Kode ini menginstal dan menggunakan pustaka nlp-id untuk mengidentifikasi dan menghapus stopword (kata-kata umum) dari teks berbahasa Indonesia. Tujuannya adalah untuk membersihkan teks sehingga hanya kata-kata yang relevan yang tersisa untuk analisis lebih lanjut.
